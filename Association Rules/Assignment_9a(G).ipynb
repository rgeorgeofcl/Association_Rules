```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(4,3,figsize=(25,15))
sns.countplot(x=books['ChildBks'],ax=ax[0][0])
sns.countplot(x=books['YouthBks'],ax=ax[0][1])
sns.countplot(x=books['CookBks'],ax=ax[0][2])
sns.countplot(x=books['DoItYBks'],ax=ax[1][0])
sns.countplot(x=books['RefBks'],ax=ax[1][1])
sns.countplot(x=books['ArtBks'],ax=ax[1][2])
sns.countplot(x=books['GeogBks'],ax=ax[2][0])
sns.countplot(x=books['ItalCook'],ax=ax[2][1])
sns.countplot(x=books['ItalAtlas'],ax=ax[2][2])
sns.countplot(x=books['ItalArt'],ax=ax[3][0])
sns.countplot(x=books['Florence'],ax=ax[3][1])
```

##**With Min Support=0.1**

```python
frequent_item_set = apriori(books, min_support=0.1, use_colnames=True)
frequent_item_set
```

##**Metric=Lift**

```python
rules=association_rules(frequent_item_set,metric='lift',min_threshold=0.6)
rules
```

```python
plt.scatter(data=rules,x='support',y='confidence')
```

```python
rules.sort_values('lift',ascending=False)
```
