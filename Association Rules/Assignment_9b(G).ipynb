## Apriori (Support=10%)

```python
frequent_item_set = apriori(movie, min_support=0.1, use_colnames=True)
frequent_item_set
```

```python
rules=association_rules(frequent_item_set,metric='lift',min_threshold=0.6)
rules
```

```python
plt.scatter(data=rules,x='support',y='confidence')
```

```python
rules.sort_values('lift',ascending=False)
```
